```python
from paddleocr import PaddleOCR, draw_ocr #main OCR dependencies
from matplotlib import pyplot as plt #plot images
import cv2 #opencv
import os #folder directory navigation
```

## 2. Instantiate Model and Detect

```python
#setup OCR
ocr_model = PaddleOCR(lang='id')
```

```python
img_path = os.path.join('.', 'indonesian drugs example.jpg')
```

```python
img_path
```

```python
#run the ocr method on the ocr model
result = ocr_model.ocr(img_path)
```

```python
result
```

```python
result[0]#where our text coordinate
```

```python
result[0][0] #box coordinate
```

```python
result[0][1][0] #first line text
```

```python
result[1]
```

```python
result[1][0] #second line
```

```python
result[1][1][0]
```

```python
for res in result:
    print(res[1][0])
```

```python
#make it in the list form
[res[1][0] for res in result]
```

## 3. Visualize Results

```python
#Extracting detected Components
boxes = [res[0] for res in result]
texts = [res[1][0] for res in result]
scores = [res[1][1] for res in result]
```

```python
boxes
```

```python
texts
```

```python
scores
```

```python
#Specifying font path for draw_ocr method
font_path = os.path.join('PaddleOCR', 'doc', 'fonts', 'latin.ttf')
```

```python
#Import our image - drug 1

#imports image
img = cv2.imread(img_path)

#reorder the color channels
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Visualize the image and detection

#resizing display area
plt.figure(figsize=(30,30))

#draw annotations on the image
annotated = draw_ocr(img, boxes, texts, scores, font_path=font_path)

#show the image using matplotlib
plt.imshow(annotated)
```
